```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import ndimage
import cv2
```

```python
cat = cv2.imread('cat.bmp',0)
dog = cv2.imread('dog.bmp',0)
```

```python
cat1 = cv2.imread('cat.bmp',0)
dog1 = cv2.imread('dog.bmp',0)
```

```python
from scipy.ndimage.filters import gaussian_filter
```

```python
cat_low = gaussian_filter(cat,6)
dog_high = dog - gaussian_filter(dog,16)
```

```python
dog_low1 = gaussian_filter(dog1,3)
cat_high1 = cat1 - gaussian_filter(cat1,4)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cd = cat_low + dog_high
plt.imshow(cd,cmap='gray')
```

```python
cd1 = dog_low1 + cat_high1
plt.imshow(cd1,cmap='gray')
```
